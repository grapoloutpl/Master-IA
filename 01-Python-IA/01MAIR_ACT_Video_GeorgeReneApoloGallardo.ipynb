## 01. Carga de datos en un DataFrame
Aquí procedo a cargar el archivo CSV con las mediciones de calidad del aire. Para ello, primero indico la ruta en la que se encuentra el fichero dentro de mi ordenador y, posteriormente, utilizo la función read_csv() de Pandas para importarlo en un DataFrame. Finalmente, muestro las primeras filas para comprobar que la carga se ha realizado correctamente.

```python
# Ruta del archivo (carpeta Descargas del usuario georgeapolo)
ruta_csv = "/Users/georgeapolo/Downloads/valencia_pollution_dataset.csv"

# Carga de datos
df = pd.read_csv(ruta_csv)

# Vista inicial
df.head()
```

## 02. Rango temporal del dataset y número de estaciones con sensor de CO
En este paso transformo la columna `fecha_carga` al tipo de dato datetime, lo que me permite trabajar correctamente con información temporal. A partir de esta conversión, calculo la fecha más antigua y la más reciente del dataset para identificar el intervalo temporal cubierto por la muestra. Además, contabilizo cuántas estaciones disponen de sensor de monóxido de carbono (co), considerando únicamente aquellas con valores no nulos.

```python
# Conversión de la columna de fecha
df["fecha_carga"] = pd.to_datetime(df["fecha_carga"], errors="coerce")

# Rango temporal
fecha_inicio = df["fecha_carga"].min()
fecha_fin = df["fecha_carga"].max()

print("Fecha de inicio del dataset:", fecha_inicio)
print("Fecha de fin del dataset   :", fecha_fin)

# Número de estaciones con sensor de CO (co no nulo)
estaciones_con_co = df[df["co"].notna()]["objectid"].nunique()
print("Número de estaciones con sensor de CO:", estaciones_con_co)
```

## 03. Estación con la media más alta de CO

En esta parte, siguiendo las indicaciones del profesor y aplicando lo aprendido en clase sobre agrupación y análisis de variables numéricas, calculo la media de los valores de monóxido de carbono (co) para cada estación.
Para ello, agrupo el DataFrame por el identificador de estación (objectid) y obtengo la media de co, ignorando posibles valores nulos. A continuación, identifico qué estación presenta la concentración media más alta y almaceno tanto su código como el valor medio correspondiente.

```python
# Media de CO por estación (ignorando valores nulos)
media_co = df.groupby("objectid")["co"].mean()

# Estación con mayor media
estacion_top_co = media_co.idxmax()
media_top = media_co.max()

print("Estación con mayor media de CO:", estacion_top_co)
print("Media de CO en esa estación   :", media_top)
```

## 04. Histograma de NO₂ de la estación con mayor media de CO
En este apartado, y siguiendo lo aprendido en clase sobre visualización de datos y análisis exploratorio, utilizo la estación que identifiqué previamente como la que presenta la mayor media de CO. A partir de ella, filtro el DataFrame para quedarme únicamente con sus registros y represento un histograma con los valores de dióxido de nitrógeno (no2). Y con este gráfico puedo observar cómo se distribuyen las concentraciones de NO₂ en dicha estación y detectar posibles patrones o variabilidad en las mediciones, para si fuera neecsario tomar acciones pertinentes 

```python
# Filtramos los datos de la estación seleccionada
df_estacion = df[df["objectid"] == estacion_top_co]

plt.figure(figsize=(6, 4))
plt.hist(df_estacion["no2"].dropna(), bins=20)
plt.title(f"Histograma NO₂ – Estación {estacion_top_co}")
plt.xlabel("NO₂ (µg/m³)")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

## 05. Serie temporal de O₃ de la estación seleccionada
En esta parte continúo con el análisis de la estación que presenta la mayor media de CO. Siguiendo las indicaciones del profesor y aplicando los conceptos trabajados en clase sobre series temporales, ordeno las mediciones de la estación por la columna fecha_carga.
De esta forma, puedo representar correctamente la evolución del ozono (o3) a lo largo del tiempo. Esta visualización me permite observar cómo varían los niveles de O₃ en esa estación durante el periodo registrado en el dataset, identificando tendencias, picos o descensos significativosy de esta manera poder sacar concluiones. 

```python
# Ordenamos por fecha para la serie temporal
df_estacion_sorted = df_estacion.sort_values("fecha_carga")

plt.figure(figsize=(7, 4))
plt.plot(df_estacion_sorted["fecha_carga"], df_estacion_sorted["o3"])
plt.title(f"Serie temporal de Ozono (O₃) – Estación {estacion_top_co}")
plt.xlabel("Fecha")
plt.ylabel("O₃ (µg/m³)")
plt.grid(True)
plt.show()
```
